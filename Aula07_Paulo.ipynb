```python
def equacao1(x,y,z):
  return 3*x + 2*y + 4*z
def equacao2(x,y,z):
  return x + y + 2*z
def equacao3(x,y,z):
  return 4*x + 3*y + 2*z

def erro(c,y):
  return abs(c-y)/c

def media(x1,x2,x3):
  return (x1+x2+x3)/3

def teste(vet, exe, tam, menor):
  for x in range (0,25):
    for i in range (0,tam):
      if (vet[exe, i] >  menor):
        menor=i
    for i in range (menor,tam):
      vet[exe, i] = vet[exe, ++i]
  return vet, tam-25
```

```python
import random
import numpy as np
repete = 1000
vezes = 5
mat = np.zeros((vezes, repete ))
for i in range (0,vezes):
  print("execução ",i)
  for x in range (0,repete):
    print("interação ",x)
    #gera números aleatórios
    x1 = round(random.uniform(-1, 1), 1)
    x2 = round(random.uniform(-2, 2), 1)
    x3 = round(random.uniform(-3, 3), 1)
    y1 = round(random.uniform(-1, 1), 1)
    y2 = round(random.uniform(-2, 2), 1)
    y3 = round(random.uniform(-3, 3), 1)
    z1 = round(random.uniform(-1, 1), 1)
    z2 = round(random.uniform(-2, 2), 1)
    z3 = round(random.uniform(-3, 3), 1)
    #calcula as equações
    eq1 = equacao1(x1,y1,z1)
    eq2 = equacao2(x2,y2,z2)
    eq3 = equacao3(x3,y3,z3)
    #calcula o percentual do erro
    e1 = (erro(1,eq1))*100
    e2 = (erro(2,eq2))*100
    e3 = (erro(3,eq3))*100
    #calcula a média do erro
    em = media(e1,e2,e3)
    
    print(e1,e2,e3," = ",em,"\n")
    #calcula    
    mat[i, x] = em
    if((x+1)%50==0):
      mat, x=teste(mat, i, x, mat[i, 0])
```
